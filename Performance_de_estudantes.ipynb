```python
provas = ['math score', 'reading score', 'writing score']
```

```python
df
```

```python
#ordenando o dataset
#a ordenação é feita pela coluna de provas,resetando o index é preciso pois quando é ordenando o index fica embaralhado,ordenando do maior p menor
df = df.sort_values(by = provas, ascending = False)\
.reset_index(drop = True)
```

```python
df
```

```python
#calcular coluna de média 
#para todas as colunas de prova,cria uma nova coluna media ,(axis=1) calcula media das linhas
df['média']=df[provas].mean(axis=1)
```

```python
df.head()
```

```python
#consultas(query)
#filtra os alunos do sexo masculino e que nao fizeram teste preparatoio e mota maior que 70,o AND é usado pois quero que todas as condiçoes sejam atendidas
#QUANDO A COLUNA TIVER ESPAÇO USAR A CRASE ´`
#DENTRO DO MEU DF COLOQUEI AS CONDIÇOES ASEREM ATENDIDAS
df.query('(gender == "male") & (`test preparation course` == "none") & (`math score` >= 70)')
```

```python
# alem de selecionar colunas,seleção boleana tbm é feita
df[(df.gender == 'male') & (df['test preparation course'] == 'none') & (df['math score'] >= 70)]
```

```python
# agrupamento - agrupa os dados por gênero e obtém estatísticas descritivas
#agrupando pela coluna genero,media e mediana por provas .T transpoe um resultado
df.groupby(by = 'gender')[provas].agg([np.mean, np.median]).T
```
