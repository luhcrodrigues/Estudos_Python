EXEMPLO

*o dataframe df1 tem as chaves "A", "B", "C" e "D", enquanto o dataframe df2 tem as chaves "B", "D", "E" e "F

```python
import pandas as pd

# Criando o primeiro dataframe
df1 = pd.DataFrame({'chave': ['A', 'B', 'C', 'D'],
                    'valor1': [1, 2, 3, 4]})

# Criando o segundo dataframe
df2 = pd.DataFrame({'chave': ['B', 'D', 'E', 'F'],
                    'valor2': [5, 6, 7, 8]})
```

```python
df1
```

```python
df2
```

* O argumento "how='left'" para indicar que queremos fazer um left merge join. Isso significa que todas as linhas do dataframe df1 sero mantidas no resultado final, e apenas as linhas do dataframe df2 que tm a mesma chave sero adicionadas.

* O resultado do left merge join  um novo dataframe com as colunas de ambos os dataframes, e linhas que correspondem s chaves que aparecem em ambos os dataframes. Se uma chave aparece apenas em um dos dataframes, o valor correspondente na outra coluna ser preenchido com NaN (not a number).

```python
# A funo recebe os dois dataframes e a coluna a ser usada como chave de juno (no nosso caso, a coluna "chave").
# Fazendo o left merge join
df_joined = pd.merge(df1, df2, on='chave', how='left')
```

```python
# Imprimindo o resultado do left merge join
print(df_joined)
```

```python
import pandas as pd

# Criando o primeiro dataframe
df1 = pd.DataFrame({'chave': ['A', 'B', 'C', 'D'],
                    'valor1': [1, 2, 3, 4]})

# Criando o segundo dataframe
df2 = pd.DataFrame({'chave': ['B', 'D', 'E', 'F'],
                    'valor2': [5, 6, 7, 8]})
```

```python
# Fazendo o right merge join
df_joined = pd.merge(df1, df2, on='chave', how='right')
```

```python
# Imprimindo o resultado do right merge join
print(df_joined)
```

```python
import pandas as pd

# Criando o primeiro dataframe
df1 = pd.DataFrame({'chave': ['A', 'B', 'C', 'D'],
                    'valor1': [1, 2, 3, 4]})

# Criando o segundo dataframe
df2 = pd.DataFrame({'chave': ['B', 'D', 'E', 'F'],
                    'valor2': [5, 6, 7, 8]})
```

```python
# Fazendo o outer merge join
df_joined = pd.merge(df1, df2, on='chave', how='outer')
```


# GROUPBY
USANDO PARA AGRUPAR POR TEMAS,COLUNAS ETC

```python
data = {'Sede':['SP','SP','MG','MG','RJ','RJ','ES','ES'],
       'Vendedor':['Jorge','Ana','Tiago','Pedro','Raphaela','Guto','Maria','Carolina'],
       'Vendas':[2000,2500,3100,1200,1500,3900,2900,3900]}
df = pd.DataFrame(data)
```

```python
df
```

```python
#agrupe por qual coluna?
df.groupby('Sede')
```

```python
by_sede=df.groupby('Sede')
```

```python
#qual sede vendeu mais
by_sede.max()
```

```python
#media
by_sede.mean()
```

```python
#descreve as estatisticas da coluna,trazendo uma viso geral
by_sede.describe()
```

```python
#transpose() troca o que  coluna p linha e vice versa
by_sede.describe().transpose()
```

```python
#TRAZ A DESCRIO APENAS INFORMAOES DE UMA COLUNA EM ESPECIFICO
by_sede.describe().transpose()['SP']
```
