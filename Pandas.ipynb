```python
import pandas as pd
import numpy as np
```

```python
#SERIES A PARTIR DE UMA LISTA
lista=[10,20,30,40]
```

```python
#metodo series e colocando no parênteses o que vai ser transformado
pd.Series(lista)
```

```python
#series a partir de um dicionario
dic={'telefone':'0000000','nome':'luana','idade':'39','qtd_filhos':'1'}
```

```python
#as chaves se tornam índices 
pd.Series(dic)
```
