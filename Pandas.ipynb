# Indexao e Seleo de df

```python
import pandas as pd
import numpy as np
```

```python
indices=['Janeiro','Maro','Maio','Julho','Agosto','Outubro','Dezembro']
```

```python
colunas=np.arange(1,32)
```

```python
colunas
```

```python
dados=np.random.randint(492,2050,size=(7,31))
```

```python
dados
```

```python
df4=pd.DataFrame(data=dados,index= indices,columns=colunas)
```

```python
df4
```

```python
#retorna a coluna 2
df4[2]
```

```python
#traz mais de uma coluna
df4[[1,2,3,4]]
```

```python
#[coluna] [linha]
df4[1][4]
```

```python
#df4[linha:linha]
df4[2:]
```

```python
#loc =localize[indice da linha :indice da coluna]#
df4.loc['Outubro',3]
```

```python
#iloc(indice numerico da linha:indice numerico da coluna)
df4.iloc[3,3]
```

Quanto vendeu os dias 10,20,30 de todos os meses

```python
#mais de uma coluna  feita uma lista
df4[[10,20,30]]
```

Quanto vendeu no dia 15 de maro?

```python
#especificando a coluna e a linha
df4[15] ['Maro']
```

Quanto vendeu no mses de  janeiro,maro,maio?

Quanto vendeu nos meses de Julho,agosto,outubro nos dias 25 a 31

```python
#mais de uma linha e colunas,inseiri listas,pesquisar pelos nomes
df4.loc[['Julho','Agosto','Outubro'],[25,26,27,28,29,30,31]] 
```

```python
#faz a busca pelos indices numericos de colunas e linhas
df4.iloc[[3,4,5],[24,25,26,27,28,29,30]]
```

```python
#renomeando 

#df.rename{nome atual:nome futuro}
```

```python
#traz em valores boleanos
df4>1000
```

```python
#df onde os valores sao maiores que mil e os que no forem ele retorna como nul
df4[df4>1000]
```

```python
#fillna faz o preenchimento de nulos por qualquer coisa que foir adicionada no parmetro 
#traz os valores so maiores que mil
df4[df4>1000].fillna('-')
```

```python
#retorna um dataframe maiores que mil e colocando as linhas que eu quero
df4[df4>1000][0:3]
```

```python
#[condio][linhas[:]traz tudo][[colunas]]
df4[df4>1000][:][[1,2,3,4,5]].fillna('-')
```

```python
#duas condioes  passado uma tupla ( E)
df4[(df4[10]>1000) & (df4[15]>1500)]
```

```python
#( OU )
df4[(df4[10]>1000) | (df4[15]>1500)]
```
