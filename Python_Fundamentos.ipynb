```python
contador = 1

while contador <= 20:#enquanto meu contador foi maior = 10
    print(contador)# print
    contador = contador + 1 # o novo contador receba o contador atual,que começou valendo 0 + 1 sendo igual a 2 
```

Break

```python
contador = 1

while contador <= 20:#enquanto meu contador foi maior = 10
    print(contador)# print
    if contador==7:#se meu contador for igual a 7 break
        break#parar
    contador = contador + 1 # o novo contador receba o contador atual,que começou valendo 0 + 1 sendo igual a 2 
```

Continue

```python
#retorna o comando do loop para o while,quando acha a condição pula a ação e volta para o loop
contador = 1
while contador < 10:#enquanto meu contador foi maior = 10
    contador = contador + 1
    if (contador==7):#se meu contador for igual a 7 break
        continue 
    print(contador)
```

# TUPLAS




1.    conter qualquer tipo de valor, incluindo números, strings, outras tuplas e objetos.
2.  ()
3. armazenam um conjunto de valores separados por vírgulas
4.A principal diferença é que as tuplas são imutáveis, o que significa que não podem ser modificadas depois de criadas.
5.As tuplas são frequentemente usadas para retornar múltiplos valores de uma função

```python
#Criando uma tupla
tupla = (1, 2, 3, "quatro")
```

```python
#Acessando elementos da tupla:
tupla = (1, 2, 3, "quatro")
print(tupla[0])  
print(tupla[3])  
```

```python
#Iterando sobre os elementos da tupla
#Iterar significa percorrer uma estrutura de dados, como uma lista, tupla, dicionário ou conjunto, para acessar e manipular cada um de seus elementos.
#podemos iterar sobre uma estrutura de dados usando um loop for. O loop for é uma estrutura de controle que nos permite executar um bloco de código para cada elemento de uma sequência
tupla = (1, 2, 3, "quatro")
for elemento in tupla:
    print(elemento)
```

```python
#Concatenando duas tuplas:
tupla1 = (1, 2, 3)
tupla2 = ("quatro", "cinco", "seis")
tupla_concatenada = tupla1 + tupla2
print(tupla_concatenada)
```

```python
#Usando tuplas como chaves de dicionários:
ponto = (3, 4)
dicionario = {ponto: "ponto de origem"}
print(dicionario)
```

```python
#loop in range(no intervalo de)
for valor in range(0,21):
  print(valor)
```

```python
#Looop em lista
lista = [10, 20, 30, 40, 50]

# Usando um loop for para percorrer a lista e imprimir cada item
for x in lista:
    print(x)#retornando cada valor do elementos 
```

```python
#loop tupla
minha_tupla = (10, 20, 30, 40, 50)

# Usando um loop for para percorrer a tupla e imprimir cada item
# a variável item assume cada valor da tupla em cada iteração do loop for.
for item in minha_tupla:
    print(item)
```

```python
#dicionarios
meu_dicionario = {"nome": "João", "idade": 30, "cidade": "São Paulo"}

# Usando um loop for para percorrer o dicionário e imprimir cada chave e valor
for chave, valor in meu_dicionario.items():#o método items() é usado para retornar uma lista de tuplas, onde cada tupla contém a chave e o valor de um par de chave-valor do dicionário
    print(chave, ":", valor)
```

```python
# percorrendo so as chave
for chave in meu_dicionario:
  print(chave)
```

```python
# percorrendo so valores
for valor in meu_dicionario.values():
  print(valor)
```

```python
#break
for valor in range (0,10):
   if valor == 7:
     break
   print(valor)

     
        
    
```

```python
#continue
for valor in range (0,21):
   if (valor == 5 or valor == 10 or valor == 15 or valor == 20 ):
     continue
   print(valor)
```

```python
salarios = [1000, 1500, 2000, 2500, 3000]
```

```python
#map(percorrer todos os valores da lista e realiza a ação para acda valor )]
#map(função,iteravel(pode passar de valor por valor fazendo alteração))
#1 funçaõ tradicional
def aumento (x):
  y=x +(x*0.25)#nova variavel recebe o valor de x + os 0.25
  return y#valor que guarda a soma 
```

```python
#o map não pode ser visualizado entao foi para uma variavel e depois transformado em lista
salarios_atual=map(aumento,salarios)
```

```python
#para aparecer foi transformado em lista com a função LIST
print(list(salarios_atual))
```

```python
#funçao lambda
salarios = [1000, 1500, 2000, 2500, 3000]
```

```python
salariio_atual=map(lambda x:x+(x*0.25),salarios)
```

```python
print(list(salariio_atual))
```

```python
# função lambda recebe um argumento e retorna o quadrado desse argumento
quadrado = lambda x: x**2
```

```python
numeros = [1, 2, 3, 4, 5]
quadrados = list(map(lambda x: x**2, numeros))
print(quadrados)
```
